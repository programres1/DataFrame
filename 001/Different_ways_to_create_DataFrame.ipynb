#Different ways to create DataFrame


```python
# Import pandas library
import pandas as pd
import numpy as np
```

**Method 1: From list of lists**

```python
# list of lists
data = [['ram', 10], ['raju', 11], ['rajesh', 14]]
 
# pass column names as list in columns parameter 
df = pd.DataFrame(data, columns = ['Name', 'Age']) 
 
print(df)

df = pd.DataFrame(data)
df
```

**Method 2: From 2-D narray**

```python
#np array
data = np.array([[2014,"toyota","corolla"], 
                 [2018,"honda","civic"], 
                 [2020,"hyndai","accent"], 
                 [2017,"nissan","sentra"]]) 


df = pd.DataFrame(data, columns = ['year', 'make','model'])
df
```

**Method 3: From dict of narray/lists**

```python
data = {
    'Name':['Tom', 'nick', 'krish', 'jack'],
        'Age':[20, 21, 19, 18]
        }

#Dictionary Keys become Column names in the dataframe 
# pass index names as list in index parameter 
df = pd.DataFrame(data,index =['rank1',
                                'rank2',
                                'rank3',
                                'rank4'])
 

print(df)


data = {
    'Name':np.array(['Tom', 'nick', 'krish', 'jack']),
        'Age':[20, 21, 19, 18]
        }
 

df = pd.DataFrame(data,index =['rank1',
                                'rank2',
                                'rank3',
                                'rank4'])
 

df
```

**Method 4:Using from_dict method**

Using from_dict, we have the ability to chose any column as an index of the dataframe(Difference from Method-3).orient='index' makes it possible.
It creates data frames as transpose, in creation phase itself.

```python
data = {'year': [2014, 2018,2020,2017], 
        'make': ["toyota", "honda","hyndai","nissan"],
        'model':["corolla", "civic","accent","sentra"]
       }

df = pd.DataFrame.from_dict(data)
print(df)

df = pd.DataFrame.from_dict(data,orient='index')
df  
```

**Method 5: From list of dicts**

```python
data = [{'a': 1, 'b': 2, 'c':3},
        {'a':10, 'b': 20, 'c': 30}]
 

df = pd.DataFrame(data)
 

df
```

**Method 6: From list of tuples**

```python
# List1
Name = ['tom', 'krish', 'nick', 'juli']
   
# List2
Age = [25, 30, 26, 22]

#List 3

Rank=[1,2,3,4]

# merge list with zip() and then cast to list.
list_of_tuples = list(zip(Name, Age,Rank))

# Converting lists of tuples into pandas Dataframe.
df = pd.DataFrame(list_of_tuples,
                  columns = ['Name', 'Age','Rank'])
    

df
```

**Method 7: From Dicts of pandas series**

```python
#List
author = ['Jitender', 'Purnima', 'Arpit', 'Jyoti']
article = [210, 211, 114, 178]

#Create pandas series from list  
auth_series = pd.Series(author)
article_series = pd.Series(article)

#Create dict from pandas series  
frame = { 'Author': auth_series, 'Article': article_series }
  
df = pd.DataFrame(frame)

 
df


#How to add series externally in dataframe

age = [21, 21, 24, 23]
  
df['Age'] = pd.Series(age)


# Print data. 
df
```

**Method 8 : From Blank dataframe**

```python
# Initialize a blank dataframe 
df = pd.DataFrame(columns = ['year','make','model'])
# Add records to dataframe using the .loc function
df.loc[0] = [2014,"toyota","corolla"] 
df.loc[1] = [2018,"honda","civic"] 
df
```
